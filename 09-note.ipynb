問106

```python
import random

while True:
 rand_num = random.randint(1,6)
 print(rand_num)
 if rand_num == 4:
    continue
 print('.')
```

4の後はピリオドが表示されない。continueがあると先を無視して繰り返しをやめて最初に戻る
